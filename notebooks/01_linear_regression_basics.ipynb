# 1. Problem Motivation

In thermal systems engineering we often want to predict relationships such as:

Ambient Temperature → Compressor Power

Instead of manually fitting equations, machine learning can learn this relationship directly from data.

# 2. Mathematical Intuition

The simplest machine learning model is **linear regression**.

The model assumes a relationship of the form:

y = wx + b

Where:

x = input variable (ambient temperature)  
y = output variable (compressor power)

w = slope of the relationship  
b = intercept

# 3. Engineering Analogy

Thermal engineers often perform similar tasks during:

• system identification  
• calibration curve fitting  
• empirical modeling

Machine learning automates this process by discovering the optimal parameters from data.

```python
# 4. Python Implementation

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

ambient_temp = np.array([25,30,35,40]).reshape(-1,1)
compressor_power = np.array([1.5,2.1,2.8,3.6])

model = LinearRegression()
model.fit(ambient_temp, compressor_power)

predictions = model.predict(ambient_temp)

plt.scatter(ambient_temp, compressor_power)
plt.plot(ambient_temp, predictions)

plt.xlabel("Ambient Temperature (°C)")
plt.ylabel("Compressor Power (kW)")
plt.title("Linear Regression Example")

plt.show()

print("Slope:", model.coef_)
print("Intercept:", model.intercept_)
```

# 5. Visualization

The regression line represents the model's learned relationship between ambient temperature and compressor power.

The algorithm determines the slope and intercept that minimize prediction error.

# 6. Key Takeaways

• Machine learning discovers relationships from data  
• Linear regression is the simplest ML model  
• Many advanced ML models are extensions of this idea
